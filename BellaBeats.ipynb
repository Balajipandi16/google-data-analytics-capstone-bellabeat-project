```python


# Assuming df11 is already defined with 'WeightKg' and 'Fat' columns
# If not, you can generate it like this (for example purposes):
np.random.seed(0)
df11 = pd.DataFrame({
    'WeightKg': np.random.uniform(50, 100, 100),  # Random weights between 50 and 100 kg
    'Fat': np.random.uniform(0.1, 0.3, 100)       # Random fat percentages between 10% and 30%
})

# Scatter Plot
plt.figure(figsize=(8, 6))
plt.scatter(df11['WeightKg'], df11['Fat'], c='blue', alpha=0.5)

plt.title('Scatter Plot of Weight vs Fat')
plt.xlabel('Weight (Kg)')
plt.ylabel('Fat (%)')

# Optionally, add a trend line to see correlation
m, b = np.polyfit(df11['WeightKg'], df11['Fat'], 1)
plt.plot(df11['WeightKg'], m*df11['WeightKg'] + b, color='red')

plt.show()
```
